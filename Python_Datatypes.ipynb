```python
#What are Datatypes?
# ans :- Datatypes tell Python what kind of data is stored.
# Main Datatypes :
# | Datatype | Example | Meaning         |
# | -------- | ------- | --------------- |
# | int      | 10      | Integer number  |
# | float    | 10.5    | Decimal number  |
# | str      | "Hello" | Text            |
# | bool     | True    | True/False      |
# | list     | [1,2,3] | Multiple values |
# | tuple    | (1,2,3) | Fixed values    |
# | dict     | {"a":1} | Key-value pair  |

#Difference Between List, Tuple, Dictionary, NumPy
# | Feature   | List        | Tuple      | Dictionary   | NumPy Array             |
# | --------- | ----------- | ---------- | ------------ | ----------------------- |
# | Symbol    | []          | ()         | {}           | array([])               |
# | Mutable   | Yes         | No         | Yes          | Yes                     |
# | Ordered   | Yes         | Yes        | Yes          | Yes                     |
# | Key-Value | No          | No         | Yes          | No                      |
# | Faster    | Medium      | Fast       | Fast         | Very Fast               |
# | Use       | Store items | Fixed data | Data mapping | Mathematical operations

# List Example
mylist = [1, 2, 3]
print(mylist)
#Tuple Example
mytuple = (1, 2, 3)
print(mytuple)
#Dictionary Example
mydict = {
    "name": "Adil",
    "city": "Pune"
}

print(mydict)
#NumPy Example
import numpy as np

arr = np.array([1, 2, 3])

print(arr)
```
